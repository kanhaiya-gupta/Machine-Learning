```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import wasserstein_distance

# randomly initialize the RNG from some platform-dependent source of entropy
#np.random.seed(None)
np.random.seed(2147483648)

# get the initial state of the RNG
st0 = np.random.get_state()[1][0]
print("Seed value: ", st0)

bins_val = 15

N1=100
mu1, sigma1 = 100, 5
mu2, sigma2 = 10, 40
X1_N1 = np.random.normal(mu1, sigma1, N1)
X2_N1 = np.random.normal(mu2, sigma2, N1)
X1 = np.concatenate([X1_N1, X2_N1])


#np.random.seed(21474)

st1 = np.random.get_state()[1][0]
print("Seed value: ", st1)

N2 = N1
mu, sigma = 100, 5
mu2, sigma2 = 10, 40
X1_N2 = np.random.normal(mu, sigma, N2)
X2_N2 = np.random.normal(mu2, sigma2, N2)
X2 = np.concatenate([X1_N2, X2_N2])


ks_test = stats.ks_2samp(X1, X2)
p_val = ks_test[1]*100
print("p-value ", p_val)
```

```python
plt.hist(X1, bins = bins_val,  density=False, histtype='bar', rwidth= 0.85, edgecolor='black', alpha=0.6, label="Seed by numpy: value 1")#: %0.0f" %st0)
plt.hist(X2, bins = bins_val,  density=False, histtype='bar', rwidth= 0.85, edgecolor='black', alpha=0.6, label="Seed by numpy: value 2")#: %0.0f" %st1)
plt.grid()
plt.legend()
plt.text(-60, 55, r"Sample size = %0.0f"%len(X1), weight="bold")
plt.text(-60, 50, r"$ X\sim N(\mu _{1}+\mu _{2}, \sigma _{1}^{2}+\sigma _{2}^{2}) $")
plt.text(-60, 45, r"$\mu _{1}, \sigma_{1}$ =%0.0f"%mu1 + ", %0.0f"%sigma1)
plt.text(-60, 40, r"$\mu _{2}, \sigma_{2}$ =%0.0f"%mu2 + ", %0.0f"%sigma2)
plt.text(-60, 35, "KS-test: p-value =%0.2f" %p_val +"%" )

plt.show()
```
