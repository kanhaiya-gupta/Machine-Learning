```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import wasserstein_distance
from scipy.optimize import curve_fit

# randomly initialize the RNG from some platform-dependent source of entropy
#np.random.seed(None)
np.random.seed(2147483648)
```

```python
# get the initial state of the RNG
st0 = np.random.get_state()[1][0]
print("Seed value: ", st0)

bins_val = 15

N1=100
mu_a1, sigma_a1 = 100, 5
mu_a2, sigma_a2 = 10, 40
X1_N1 = np.random.normal(mu_a1, sigma_a1, N1)
X2_N1 = np.random.normal(mu_a2, sigma_a2, N1)
X1_Ns1 = np.concatenate([X1_N1, X2_N1])


#np.random.seed(21474)

np.random.seed(2147483648)

N2 = N1
mu_b1, sigma_b1 = 100, 5
mu_b2, sigma_b2 = 10, 40
X1_N2 = np.random.normal(mu_b1, sigma_b1, N2)
X2_N2 = np.random.normal(mu_b2, sigma_b2, N2)
X2_Ns1 = np.concatenate([X1_N2, X2_N2])

ks_test_X1 = stats.ks_2samp(X1_N1, X1_N2)
p_val_X1 = ks_test_X1[1]*100

ks_test_X2 = stats.ks_2samp(X2_N1, X2_N2)
p_val_X2 = ks_test_X2[1]*100

ks_test = stats.ks_2samp(X1_Ns1, X2_Ns1)
p_val = ks_test[1]*100


dist_01 = wasserstein_distance(X1_Ns1, X2_Ns1)

print("p-value ", p_val)


st1 = np.random.get_state()[1][0]
print("Seed value: ", st1)
```
