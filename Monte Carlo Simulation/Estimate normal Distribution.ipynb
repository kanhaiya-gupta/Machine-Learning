```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
n_simulations = 100000

# Pdf to be estimated 

def Log_normal_pdf(X, params):
    # Unpack the model parameters
    
    rm, sigma = params
    
    # Compute the pdf-value
    pdf_value = (1.0/(np.sqrt(2*np.pi)*X*sigma)*np.exp(-(np.log(X/rm))**2/(2*sigma**2)))
    
    return pdf_value


# mu & sigma = 2.1 & 0.4
s = random.uniform(0, 1,  n_simulations)

mu = 2.1
sigma = 0.4
parms = [mu, sigma]

samp = s 
```

```python
xmin = 0
xmax = 10
pmax = 1.0
N_MC = 100000

t = np.random.uniform(xmin,xmax,N_MC)  #get uniform temporary x values
y = np.random.uniform(0,pmax,N_MC)  # get uniform random y values
```

```python
# plot all the t-y pairs
plt.scatter(t,y, s=0.1, c='orange')
plt.show()
```

```python
#make a mask that keeps index to the accepted pairs. Plot them
mask = y<=Log_normal_pdf(t, parms)
plt.scatter(t[mask],y[mask], s=0.1)
plt.show()
```

```python
#histogram the t values with and without the mask
plt.hist(t, bins=100)
plt.hist(t[mask], bins=100)
plt.show()
```
